```python
def calc (a,b,c):
    if c == '+':
        sum = a+b
        print (f"The sum of the numbere is {sum}")
    elif c == '-':
        dif = a-b
        print(f'The difference between the numbers is {dif}')
    elif c == '*':
        mul = a*b
        print (f'The product of the numbers is {mul}')
    elif c =='/':
        dif = a/b
        print(f'The division of the number is {dif}')

    else:
        print('Invalid operator please use '+' or '-' or '*' or '/'')
   
    
```

```python
calc(3,4,"+")
```

```python
calc (45,5,'/')
```
