### Write a function that takes from the user a FILE NAME , the user might sent a wrong file name if so it should not break. it should simply print THE CONTENT OF THE FILE OUT

```python
def read_file():
    filename = input("Please enter the file name: ")
    try:
        with open(filename, "r") as file:
            content = file.read()
            print("The content of the file is:\n")
            print(content)
    except FileNotFoundError:
        print("The file name you entered does not exist.")
    except Exception as e:
        print(f"An error occurred: {e}")

```

```python
read_file()
```
