### 9. Write a function that takes a list of numbers and returns their average.

```python
def averg(numbers):
    if len(numbers) == 0:
        return 0  
    total = 0
    for num in numbers:
        total += num
    return total / len(numbers)
```

```python
averg([32,55555,67,8])
```

### 10. Write a function that counts vowels in a string.

```python
def vowel_count(s):
    vowels = "aeiouAEIOU"    
    count = 0
    for char in s:
        if char in vowels:
            count += 1
    return count
```

### 11. Write a function that takes a list and returns a new list with unique values.

```python
def uni_val(list):
    unique_list = []
    for item in list:
        if item not in unique_list:
            unique_list.append(item)
    return unique_list
```

```python
num = [1,2,3,4,1,1,22,3]
result = uni_val(num)
print(result)
```

### 12. Write a function that checks if a word is a palindrome.

```python
def check (word):
    word = word.lower()

    if word == word[::-1]:
        return ('The word is a palindrome')
    else:
        return ("The word is not a palindrome")
```

```python
check('lawal')
```

### 15. Write a function that returns the factorial of a number.

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```

```python
factorial(6)
```

### 17. Write a function that returns the minimum value in a list.

```python
def find_min(list):
   
    minimum = list[0]   # start with the first element
    for item in list:
   
        if item < minimum:
            minimum = item
    return minimum
```

```python
x = [2,3,4,5]
find_min(x)
```

### 18. Write a function that converts minutes to hours and minutes.

```python
def conv_min(total_minutes):
    
    hours = total_minutes // 60   # integer division gives whole hours
    minutes = total_minutes % 60  # remainder gives leftover minutes
    return (f'{total_minutes} minnutes is equivalent to {hours} hours and {minutes} minutes')
```

```python
conv_min(455)
```

### 25. Write a function that accepts score and returns grade using default grading scale.

```python
def grade(score):
   
    if score >= 70:
        return "A"
    elif score >= 60:
        return "B"
    elif score >= 50:
        return "C"
    elif score >= 40:
        return "D"
    else:
        return "F"
```

###  27. Write a function that checks if a number is prime.

```python
def is_prime(num):
   
    if num <= 1:
        return 'The number is not prime'
    for i in range(2, int(num**0.5) + 1):  
        if num % i == 0:
            return 'The number is not prime'
    return 'The number is not prime'
```

```python
is_prime(44)
```

### 28. Write a function that returns the sum of digits of a number

```python
def sum_of_digits(num):
    
    total = 0
    for digit in str(num):   # convert to string, handle negative numbers
        total += int(digit)
    return total
```

### 29. Write a function that accepts a list of prices and applies a default discount.

```python
def disc(prices, discount=0.10):

    discounted_prices = []
    for price in prices:
        new_price = price - (price * discount)
        discounted_prices.append(new_price)
    return discounted_prices
```

```python
u = [230,450]

disc(u)
```

### 30. Write a function that formats and returns a full name from first and last name.

```python
def full_name(first_name, last_name):
    
    first = first_name.strip().capitalize()
    last = last_name.strip().capitalize()
    return f"{first} {last}"
```

```python
full_name('hala','madrid')
```
